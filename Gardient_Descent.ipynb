```python
import numpy as np

# y=mx+b   #m slope #b intercept
#cal m derivative
#cal b derivative
#learning rate

def grad_des(x,y):
  pass
  m_curr= b_curr=0
  itr=10000  #trailnderror
  n=len(x)
  learning_rate=0.08   #trailnderror
  for i in range(itr):
    y_pred=m_curr * x + b_curr                         #linear equation
    cost=(1/n)*sum([val**2 for val in (y-y_pred)])     #MSE
    md=-(2/n)*sum(x*(y-y_pred))                        #Partial Derv of m(slope)
    bd=-(2/n)*sum(y-y_pred)                        #Partial Derv of b(intercept)
    m_curr=m_curr-learning_rate*md                 #adjusting with Learning rate
    b_curr=b_curr-learning_rate*bd                 #adjusting with Learning rate
    print("m{},b{},cost{} ,iteration{},".format(m_curr,b_curr,cost,i))


x=np.array([1,2,3,4,5])
y=np.array([5,7,9,11,13])
```

```python
grad_des(x,y)
```
